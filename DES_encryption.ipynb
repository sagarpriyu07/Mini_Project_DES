```python
# !pip install pycryptodome
# Import libraries
from Crypto.Cipher import DES
from google.colab import files
import ipywidgets as widgets
from IPython.display import display, HTML

# Helper function: Pad text to be a multiple of 8 bytes
def pad_text(text):
    padding_len = 8 - (len(text) % 8)
    return text + bytes([padding_len] * padding_len)  # Add padding as bytes

# Helper function: Unpad text to remove padding after decryption
def unpad_text(padded_text):
    padding_len = padded_text[-1]  # Get the last byte to determine padding length
    return padded_text[:-padding_len]  # Remove padding

# Encrypt the PDF file
def encrypt_pdf(input_file, output_file, key):
    # Ensure the key is 8 bytes long
    if len(key) != 8:
        raise ValueError("Key must be exactly 8 bytes long.")

    # Create DES cipher
    des = DES.new(key, DES.MODE_ECB)

    with open(input_file, 'rb') as f:
        plaintext = f.read()

    # Pad plaintext and encrypt
    padded_text = pad_text(plaintext)
    ciphertext = des.encrypt(padded_text)

    # Write encrypted data to output file
    with open(output_file, 'wb') as f:
        f.write(ciphertext)

    print(f"PDF file '{input_file}' encrypted and saved as '{output_file}'.")

# Decrypt the PDF file
def decrypt_pdf(input_file, output_file, key):
    # Ensure the key is 8 bytes long
    if len(key) != 8:
        raise ValueError("Key must be exactly 8 bytes long.")

    # Create DES cipher
    des = DES.new(key, DES.MODE_ECB)

    with open(input_file, 'rb') as f:
        ciphertext = f.read()

    # Decrypt ciphertext
    decrypted_data = des.decrypt(ciphertext)
    unpadded_data = unpad_text(decrypted_data)  # Remove padding

    # Write decrypted data to output file
    with open(output_file, 'wb') as f:
        f.write(unpadded_data)

    print(f"Encrypted PDF file '{input_file}' decrypted and saved as '{output_file}'.")

# Add custom CSS for styling
def custom_css():
    display(HTML('''
    <style>
        .widget-label {
            font-size: 18px;
            font-weight: bold;
            color: #4CAF50;
        }
        .widget-button {
            background-color: #4CAF50;
            color: white;
            font-size: 16px;
            padding: 10px 20px;
            border-radius: 5px;
            cursor: pointer;
        }
        .widget-button:hover {
            background-color: #45a049;
        }
        .widget-dropdown, .widget-text, .widget-password {
            width: 60%;
            margin: 10px 0;
        }
        .widget-file-upload {
            margin: 10px 0;
        }
        .header {
            font-size: 24px;
            font-weight: bold;
            color: #333;
            text-align: center;
            margin-bottom: 20px;
        }
        .description {
            font-size: 16px;
            color: #666;
            text-align: center;
            margin-bottom: 30px;
        }
    </style>
    '''))

# Frontend interface with widgets
def frontend_interface():
    custom_css()  # Apply custom styles

    # Header section
    header = widgets.HTML('<div class="header">Welcome to the DES PDF Encryptor/Decryptor</div>')
    description = widgets.HTML('<div class="description">Upload your PDF, choose an operation (Encrypt or Decrypt), and provide a key to proceed.</div>')

    # File upload widget
    upload_button = widgets.FileUpload(
        accept='.pdf',
        multiple=False,
        description='Upload PDF',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='80%')
    )

    # Key input widget
    key_input = widgets.Password(
        description='Key (8 chars):',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='80%')
    )

    # Operation choice widget
    operation_choice = widgets.Dropdown(
        options=['Encrypt', 'Decrypt'],
        description='Operation:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='80%')
    )

    # Output file name input
    output_file_input = widgets.Text(
        description='Output File Name:',
        style={'description_width': 'initial'},
        layout=widgets.Layout(width='80%')
    )

    # Action button
    action_button = widgets.Button(description="Run", button_style='success', layout=widgets.Layout(width='80%'))

    # Display all widgets
    display(header)
    display(description)
    display(upload_button)
    display(key_input)
    display(operation_choice)
    display(output_file_input)
    display(action_button)

    # Button click event handler
    def on_button_click(b):
        # Get inputs from the widgets
        if not upload_button.value:
            print("Please upload a PDF file.")
            return

        input_file = list(upload_button.value.keys())[0]
        output_file = output_file_input.value
        key = key_input.value.encode('utf-8')
        operation = operation_choice.value

        if len(key) != 8:
            print("Key must be exactly 8 characters long.")
            return

        try:
            if operation == "Encrypt":
                encrypt_pdf(input_file, output_file, key)
            elif operation == "Decrypt":
                decrypt_pdf(input_file, output_file, key)
        except Exception as e:
            print(f"An error occurred: {e}")

    # Link the button click event to the handler
    action_button.on_click(on_button_click)

# Run the frontend interface
frontend_interface()
```
